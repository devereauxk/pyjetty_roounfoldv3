```python
def get_args_from_settings(ssettings):
    sys.argv=[' '] + ssettings.split()
    parser = argparse.ArgumentParser(description='pythia8 fastjet on the fly')
    pyconf.add_standard_pythia_args(parser)
    parser.add_argument('--output', default="test_ang_ue.root", type=str)
    parser.add_argument('--user-seed', help='pythia seed', default=1111, type=int)
    args = parser.parse_args()
    return args
```

```python
mycfg = []
ssettings = "--py-ecm 5000 --user-seed=100000 --nev 1000"
args = get_args_from_settings(ssettings)
pythia_hard = pyconf.create_and_init_pythia_from_args(args, mycfg)
```

```python
max_eta_hadron=1
parts_selector_h = fj.SelectorAbsEtaMax(max_eta_hadron)
jet_R0 = 0.4
jet_selector = fj.SelectorPtMin(80.0) & fj.SelectorPtMax(100.0) & fj.SelectorAbsEtaMax(max_eta_hadron - 1.05 * jet_R0)
n_pileup = 1
jet_selector_bg = fj.SelectorAbsEtaMax(max_eta_hadron - 1.05 + jet_R0)
jet_selector_bg_sub = fj.SelectorAbsEtaMax(max_eta_hadron - 1.05 + jet_R0)
```

```python
from pyjetty.mputils import BoltzmannEvent
be = BoltzmannEvent(mean_pt=0.7, multiplicity=2000 * max_eta_hadron * 2, max_eta=max_eta_hadron, max_pt=100)
print(be)
from pyjetty.mputils import CEventSubtractor, CSubtractorJetByJet
cs = CEventSubtractor(alpha=0, max_distance=0.25, max_eta=max_eta_hadron, bge_rho_grid_size=0.25, max_pt_correct=100)
print(cs)
```

```python
def print_jet_constits(j):
    print('[jet]', j.perp(), j.phi(), j.eta(), ':', [_c.user_index() for _c in j.constituents()])
        
```

```python
# for n in tqdm(range(args.nev)):
def next_event():
    while (1):
        if not pythia_hard.next():
            continue
        parts_pythia_h = pythiafjext.vectorize_select(pythia_hard, [pythiafjext.kFinal], 0, False)
        parts_pythia_h_selected = parts_selector_h(parts_pythia_h)

        mult_hard = len(parts_pythia_h_selected)

        jets_h = fj.sorted_by_pt(jet_selector(jet_def(parts_pythia_h_selected)))

        if len(jets_h) < 1:
            continue

        j = jets_h[0]

        #make the subjets
        subjets = fj.sorted_by_pt(sj_def(j.constituents()))
        if len(subjets) < 3:
            continue

        bg_parts = be.generate(offset=10000)
        full_event = bg_parts
        _tmp = [full_event.push_back(psj) for psj in j.constituents()]

        print('nparts in the jet', len(j.constituents()), 'nparts in the hybrid event', full_event.size())

        jets_hbg = fj.sorted_by_pt(jet_selector_bg(jet_def_bg(full_event)))        
        jbg_selected = None
        for jbg in jets_hbg:
            mpt = fjtools.matched_pt(jbg, j)
            if mpt < 0.5:
                continue
            jbg_selected = jbg
        if jbg_selected is None:
            print('no bg match')
            continue

        cs_parts = cs.process_event(full_event)
        jets_hbg_sub = fj.sorted_by_pt(jet_selector_bg_sub(jet_def_bg_sub(cs_parts)))        
        jbg_selected_sub = None
        for jbg in jets_hbg_sub:
            mpt = fjtools.matched_pt(jbg, j)
            if mpt < 0.5:
                continue
            jbg_selected_sub = jbg
        if jbg_selected_sub is None:
            print('no bg sub match')
            continue

        # return j, subjets, jbg
        return j, jbg_selected, jbg_selected_sub
```

```python
def draw_jet(j):
    # for the jet but not subjets
    pts = [p.perp() for p in fj.sorted_by_pt(j.constituents())]
    ys = [j.rapidity() - p.rapidity() for p in fj.sorted_by_pt(j.constituents())]
    phis = [j.delta_phi_to(p) for p in fj.sorted_by_pt(j.constituents())]
    
    phis.append(0.4)
    phis.append(-0.4)
    ys.append(0.4)
    ys.append(-0.4)
    pts.append(0)
    pts.append(0)
    zs = [pt/j.perp() for pt in pts]
    zs_sized = [z*1000. for z in zs]
    cs = [int(z*100.) for z in zs]

    plt.figure()
    # plt.scatter(phis, ys, c=colors, s=zs_sized, alpha=0.4, cmap="PuOr") #cmap='viridis')
    plt.scatter(phis, ys, c=cs, s=zs_sized, alpha=0.4, cmap="magma") #cmap='viridis')
    
    plt.xlabel('phi')
    plt.ylabel('y')
    plt.colorbar();  # show color scale
    plt.show()
    #plt.rcdefaults()
```

```python
def draw_subjets(j, scale_pt=0):
    sj_def = fj.JetDefinition(fj.antikt_algorithm, 0.1)
    sjs = fj.sorted_by_pt(sj_def(j.constituents()))
    # for subjets
    pts = []
    ys = []
    phis = []
    cs = []
    for i,sj in enumerate(sjs):
        sc = fj.sorted_by_pt(sj.constituents())
        pts.extend([p.perp() for p in sc])
        ys.extend([j.rapidity() - p.rapidity() for p in sc])
        phis.extend([j.delta_phi_to(p) for p in sc])
        if i == 0:
            for p in sc:
                cs.append([1,0,0,0.3])
        else:
            gr_scale = (len(sjs)-i)/len(sjs)
            gr_col = 0.4-0.3*gr_scale
            # print(gr_col)
            for p in sc:
                cs.append([gr_col, gr_col, gr_col, 0.9])
    
    phis.append(0.4)
    phis.append(-0.4)
    ys.append(0.4)
    ys.append(-0.4)
    pts.append(0)
    pts.append(0)
    if scale_pt == 0:
        scale_pt = j.perp()
    zs = [pt/scale_pt for pt in pts]
    zs_sized = [z*1000. for z in zs]
    cs.append([0,0,0,0])
    cs.append([0,0,0,0])

    plt.figure()
    # plt.scatter(phis, ys, c=colors, s=zs_sized, alpha=0.4, cmap="PuOr") #cmap='viridis')
    # plt.scatter(phis, ys, c=cs, s=zs_sized, alpha=0.4, cmap="magma") #cmap='viridis')
    plt.scatter(phis, ys, c=cs, s=zs_sized, alpha=0.4, cmap="magma") #cmap='viridis')
    
    plt.xlabel(r"$\Delta\varphi$")
    plt.ylabel(r"$\Delta y$")
    # plt.colorbar();  # show color scale
    plt.show()
    print(j.perp(), sjs[0].perp(), sjs[0].perp()/j.perp())
    # plt.rcdefaults()
```

```python
def draw_sd(j, scale_pt=0):
    sd01 = fjcontrib.SoftDrop(0, 0.1, 1.0)
    j_sd01 = sd01.result(j)
    sd01_info = fjcontrib.get_SD_jet_info(j_sd01)

    pe1 = fj.PseudoJet()
    pe2 = fj.PseudoJet()
    has_parents = j_sd01.has_parents(pe1, pe2)
    if not has_parents:
        print('[w] no parents for soft drop?')
        return
    
    # for subjets
    pts = []
    ys = []
    phis = []
    cs = []
    ncs = len(j.constituents())
    for i,sj in enumerate([j, pe1, pe2]):
        sc = fj.sorted_by_pt(sj.constituents())
        pts.extend([p.perp() for p in sc])
        ys.extend([j.rapidity() - p.rapidity() for p in sc])
        phis.extend([j.delta_phi_to(p) for p in sc])
        if i == 1:
            for p in sc:
                cs.append([1,0,0,0.3])
            continue
        if i == 2:
            for p in sc:
                cs.append([0,0,1,0.3])
            continue
        gr_scale = (ncs-i)/ncs
        gr_col = 0.4-0.3*gr_scale
        # print(gr_col)
        for p in sc:
            cs.append([gr_col, gr_col, gr_col, 0.9])
    
    phis.append(0.4)
    phis.append(-0.4)
    ys.append(0.4)
    ys.append(-0.4)
    pts.append(0)
    pts.append(0)
    if scale_pt == 0:
        scale_pt = j.perp()
    zs = [pt/scale_pt for pt in pts]
    zs_sized = [z*1000. for z in zs]
    cs.append([0,0,0,0])
    cs.append([0,0,0,0])

    plt.figure()
    # plt.scatter(phis, ys, c=colors, s=zs_sized, alpha=0.4, cmap="PuOr") #cmap='viridis')
    # plt.scatter(phis, ys, c=cs, s=zs_sized, alpha=0.4, cmap="magma") #cmap='viridis')
    plt.scatter(phis, ys, c=cs, s=zs_sized, alpha=0.4, cmap="magma") #cmap='viridis')
    
    plt.xlabel(r"$\Delta\varphi$")
    plt.ylabel(r"$\Delta y$")
    # plt.colorbar();  # show color scale
    plt.show()
    # print(j.perp(), sjs[0].perp(), sjs[0].perp()/j.perp())
    # plt.rcdefaults()
```

```python
for n in range(5):
    j, jbg, jbgsub = next_event()
    # draw_jet(j)

    draw_subjets(j, jbg.perp()/4.)
    draw_subjets(jbg, jbg.perp()/4.)
    draw_subjets(jbgsub, jbg.perp()/4.)

    draw_sd(j, jbg.perp()/4.)
    draw_sd(jbg, jbg.perp()/4.)
    draw_sd(jbgsub, jbg.perp()/4.)

    # print ('[event] ', j.perp(), j, len(j.constituents()), len(sjs), [sj.perp() for sj in sjs])
    # print ('[event] w/bg ', jbg.perp(), jbg, len(jbg.constituents()), len(sjs_bg), [sj.perp() for sj in sjs_bg])
    # print (' ')
```
